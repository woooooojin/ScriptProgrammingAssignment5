 ##### [2번 문제] 
 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.


- 다형성은 상속 관계에서 다른 클래스의 인스턴스들이 같은 멤버함수를 호출하였을 때 각각 다르게 반응하는 것이다. 즉, 슈퍼 클래스에서 인스턴스의 기본적인 틀을 제공하고 이를 상속받은 서브클래스들이 각각 다른 상태를 가지고 생성되었을 때 멤버함수를 호출하면 서브클래스들 마다 다른 결과를 반환하는 것을 다형성이라 한다.

##### 코드예시

```python
class Duck:             ##오리 클래스
    def quack(self):    ##오리 울음소리
        print("꽥꽥")
    def fly(self):
        print("날아 다닐 수 없다.")

class MallardDuck(Duck):         ##청동오리 클래스
    def fly(self):               ##울음소리는 같지만 하늘을 날 수 있다.
        print("날개로 난다.")
        
class RubberDuck(Duck):          ##장난감 오리 클래스
    def quack(self):             ##울음소리가 다를 수 있다.
        print("삑삑")
        
for duck in (MallardDuck(), RubberDuck()):  ##메소드의 이름은 같지만 결과가 다르다
    duck.quack()
    duck.fly()
    print()
```

 ##### [3번 문제] 
 
 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)
   - 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
     - count: Counter 인스턴스가 지니는 초기 정수 값
     - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)
 
     > \>\>\> c = Counter(10) <br/>
     > \>\>\> d = Counter(10, 2) <br/>
     
   - 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 \_\_str\_\_() 메소드를 Counter 클래스 내에 구현하시오.
     > \>\>\> print(c)  <br/>
     > [Count (step: 1)] 10 <br/> 
     > \>\>\> print(d)  <br/>
     > [Count (step: 2)] 10 <br/>
     
   - 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.
     > \>\>\> c.incr()  <br/>
     > \>\>\> d.incr()  <br/>
     > \>\>\> print(c)  <br/>
     > [Count (step: 1)] 11 <br/> 
     > \>\>\> print(d)  <br/>
     > [Count (step: 2)] 12 <br/>
     
   - 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오. 
     > \>\>\> c()  <br/>
     > \>\>\> d()  <br/>
     > \>\>\> print(c)  <br/>
     > [Count (step: 1)] 12 <br/> 
     > \>\>\> print(d)  <br/>
     > [Count (step: 2)] 14 <br/>
     
   - 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
     > \>\>\> c = c + 5  <br/>
     > \>\>\> d = d - 5  <br/>
     > \>\>\> print(c)  <br/>
     > [Count (step: 1)] 17 <br/> 
     > \>\>\> print(d)  <br/>
     > [Count (step: 2)] 9 <br/>  
     
   - 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
     > \>\>\> print(c) > 10  <br/>
     > True <br/> 
     > \>\>\> print(d) > 10 <br/>
     > False <br/>
     > \>\>\> print(c) < 10  <br/>
     > False <br/> 
     > \>\>\> print(d) < 10 <br/>
     > True <br/>
     > \>\>\> print(c) == 17  <br/>
     > True <br/> 
     > \>\>\> print(d) != 9 <br/>
     > False <br/>

```python
class Counter():
    def __init__(self,count,step=1):  ##생성자를 통한 인스턴스 멤버 생성
        self.count = count
        self.step = step
    def __str__(self):
        return ('[Count(step:%d)]%d' % (self.step, self.count))
    def incr(self):
        self.count += self.step
    def __call__(self):             ##인스턴스 객체를 callable가능하도록
        self.incr()
    def __add__(self,other):        ##count값에 대한 덧셈연산 후 인스턴스 그 자체를 반환
        self.count += other
        return self                 
    def __sub__(self,other):        ##count값에 대한 뺄셈연산 후 인스턴스 그 자체를 반환
        self.count -= other
        return self
    def __gt__(self,other):         ##count값에 대한 비교연산 결과 값 반환
        return self.count > other
    def __lt__(self,other):
        return self.count < other
    def __eq__(self,other):
        return self.count == other

c = Counter(10) 
d = Counter(10, 2) 
print(c)
print(d)
print()

c.incr()
d.incr()
print(c)
print(d)
print()

c()
d()
print(c)
print(d)
print()

c = c + 5 
d = d - 5 
print(c)
print(d)
print()

print (c > 10)
print (d > 10)  
print (c < 10) 
print (d < 10)  
print (c == 17) 
print (d != 9) 
```

 ##### [4번 문제] 
 
 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init\_\_(), \_\_str()\_\_(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
   - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

- \__init__() 메소드
  - MySet클래스는 list클래스를 상속받은 서브클래스로 \__init__(self,ㅣ)메소드는 list클래스의 생성자 메소드를 오버라이딩한 메소드다. \__init__(self,ㅣ)메소드는 인자로 전달받은 자료형 인스턴스의 각 element를 자기자신에게 append()하고 elimicate_duplicate()메소드를 호출하여 자기자신을 인자로 넘긴다.

- \__str__() 메소드
  - \__str__(self,ㅣ)메소드 또한 list클래스의 \__str__() 메소드를 오버라이딩한 메소드다. 즉, print()내장함수를 통해 MySet 인스턴스의 내용을 사용자가 볼 수 있도록 문자열로 나타내주는 메소드이다. MySet 인스턴스의 내용을 확인하기 위해 정해진 형식에 따라 자기자신의 element를 문자열로 나타내어 반환한다.

- elimicate_duplicate() 메소드
  - 집합 자료형은 중복된 element가 없는 자료형이다. 즉, elimicate_duplicate() 메소드는 element의 중복된 요소를 제거하는 정적 메소드다. 생성자에서 전달받은 자기자신의 각 element를 새로운 리스트에 추가하는데, 이 과정에서 이미 추가된 값이 존재하는지 검사하고 존재하지 않는 경우에만 추가한다. 이후에 자기자신을 빈메소드로 초기화하고 중복된 값이 제거된 리스트의 각 element를 자기자신에게 append함으로써 중복된 값을 제거한다.

 ##### [5번 문제] 
 
 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오 
   - 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
   - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
     > \>\>\> u = s | t  <br/>
     > \>\>\> print(u)  <br/>
     > MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} <br/> 
     
   - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
     > \>\>\> u = s & t  <br/>
     > \>\>\> print(u)  <br/>
     > MySet: {2, 3} <br/>
 
   - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
     > \>\>\> s = MySet([1, 2, 3])  <br/>
     > \>\>\> t = MySet([3, 4, 5])  <br/>
     > \>\>\> u = s - t  <br/>
     > \>\>\> print(u)  <br/>
     > MySet: {1, 2} <br/>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    def __or__(self,other):   ##합집합 연산 메소드
        result = self + other  ##피연산자 두 집합을 합친다.
        MySet.eliminate_duplicate(result)  ##중복을 제거한다.
        return MySet(result)    ##새로운 집합 인스턴스를 반환
    
    def __and__(self,other): ##교집합 연산 메소드
        result = [e for e in self if e in other] ##self의 element중 other에도 포함된 element를 원소로 하는 리스트
        return MySet(result)  ##새로운 집합 인스턴스를 반환
    
    def __sub__(self,other):
        result = [e for e in self if e not in other] ##self의 element중 other에 포함되지 않은 element를 원소로 하는 리스트
        return MySet(result)   ##새로운 집합 인스턴스를 반환

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    
    u = s | t 
    print(u)
    
    u = s & t
    print(u)
    
    s = MySet([1, 2, 3]) 
    t = MySet([3, 4, 5]) 
    u = s - t 
    print(u)  
```

     - 파라미터로 받은 단어와 유사도가 동일한 URL이 여러 개이면 해당 URL을 모두 출력하시오.  

```python
import requests
import re
import string

class SearchEngine :
    def __init__(self, *args):  ##생성자에서 인자로받은 url을 
        self.urls = []
        for url in args : self.urls.append(url) #인스턴스 멤버로 유지
    def addUrl(self,url):
        self.urls.append(url)
    def removeUrl(self,url):
        self.urls.remove(url)
    def listUrls(self):
        for url in self.urls:
            print (url)
    def getWordsFrequency(self):  ##URL사이트의 종합적인 빈도수 단어사전 반환 메소드
        if self.urls == [] : return {}  ##등록된 URL이 없으면 빈 사전 반환
        
        self.dicList = [] ##각 URL을 분석한 빈도수 사전을 저장할 리스트
        for url in self.urls:
            req = requests.get(url)
            text = SearchEngine.extractText(req.text)   #순수 텍스트 추출
            text = SearchEngine.filterSpecialChar(text) #특수문자 제거
            l = text.split()
            dic = SearchEngine.freqOfWord(l)           #출현한 단어의 빈도수 사전
            dic = SearchEngine.filterAbbreviation(dic) #사전 불용어 제거
            self.dicList.append(dic)                   #URL사전 저장
            
        self.totalDic = {}  ##종합 단어 출현빈도 사전
        for dic in self.dicList:            ##빈도사전을 하나씩 가져와서
            keys = list(dic.keys())         ##키를 리스트로 만들고
            for key in keys:                 
                if key not in self.totalDic:     ##중복된 키가 없으면 사전에 추가
                    self.totalDic[key] = dic[key]
        return self.totalDic
    
    def getMaxFreqencyWords(self):  ##종합적인 빈도수 사전에서 많이 출현한 단어 리스트 반환 메소드
        if self.urls == [] : return  ##등록된 URL이 없으면 None반환
        
        self.getWordsFrequency()  ##종합적인 빈도수 사전을 만들고
        items = list(self.totalDic.items()) 
        maxFreq = max(items,key=lambda x: x[1])[1]  ##가장 큰 빈도수를 찾는다.
        
        for word,freq in items :   ##빈도수 사전의 item중
            if freq == maxFreq : print(word) ##가장 큰 빈도수와 같은 빈도수를 가진 단어를 모두 출력
                
    def searchUrlByWord(self,keyword):
        if self.urls == [] : return  ##등록된 URL이 없으면 None반환
        
        self.getWordsFrequency() ##각 URL 사이트의 빈도수 사전 리스트를 만듬
        similarList = []  ##사이트 별 유사도를 저장할 리스트
        for dic in self.dicList:
            totalWord = 0 ##사전의 전체 출현단어 빈도수 
            for word,freq in dic.items(): totalWord += freq 
            
            similarWord = 0 ##사전에서 검색어와 유사한 단어 빈도수
            if keyword.isalpha() : keyword = keyword.lower() ##검색어가 영어이면 소문자로 변경
            for word,freq in dic.items():
                if word.isalpha() : word = word.lower()    ##빈도수 사전의 단어가 영문이면 소문자로 변경
                if keyword in word :     ##검색어 문자열이 빈도수 사전 단어 문자열에 존재하면
                        similarWord+=freq    ##유사 단어 빈도수에 사전 단어 빈도수를 더함

            similarity = similarWord/totalWord ##유사도 = 유사 단어 빈도수/전체 단어 빈도수            
            similarList.append(similarity)
        
        maxSimilar = max(similarList)   ##가장 큰 유사도를 찾아서     
        for index in range(len(similarList)):   ## Max유사도와 같은 모든 URL을 출력
            if similarList[index] == maxSimilar:
                print(self.urls[index])
        
    @staticmethod
    def extractText(source):
        text = "" #순수 텍스트 식별자
        while(1) :      
            source = source[source.find('<'):] #'<'를 찾는다
            if (source.find("<!--") == 0) : 
                    source = source[source.find("-->")+3:] #주석 내용 건너뜀 
            elif (source.find("<script") == 0) :        
                    source = source[source.find("</script>")+9:]  #자바스크립트 내용 건너뜀   
            elif (source.find("<style") == 0) :
                    source = source[source.find("</style>")+9:] #스타일 내용 건너뜀
            else :       
                source = source[source.find('>')+1:] #'>'를 찾는다
                for ch in source :    #'>' 이후부터 한글자씩 가져온다
                    if ch == '<' : break    # 다시 '<' 문자 등장하면 종료
                    elif ch == '\t' or ch == '\n' : continue    #탭,개행 문자는 순수 텍스트에 포함x
                    else : text += ch   #순수 텍스트 추가     
                if text.endswith(" ") == False : text += " " #공백으로 분리하기 위해 순수 텍스트 추출 후 뒤에 공백문자 추가
            if(source.find('<') == -1):break #'<' 문자가 더 이상 없는 경우 무한루프종료 
        return text
    
    @staticmethod
    def replaceAll(text, dic):  ##문자열을 사전에 맵핑된 문자로 대체로 하는 함수
        for i,j in dic.items():
            text = text.replace(i,j,text.count(i)) #문자열에 사전에 키와 같은 문자가 존재하면 키에 맵핑된 value값으로 대체한다.
        return text  
    
    @staticmethod
    def filterSpecialChar(text):
        keys = [ch for ch in string.punctuation] #구두문자를 key리스트로 만듬
        #구두문자 이외의 특수문자를 key리스트에 추가
        keys = keys + ["★","☆","※","●","nbsp","lt","amp","quot","middot","uarr","rarr","darr","harr","💕",
                       "✒✒","（╹◡╹）","♥","🛫","\u200b"]
    
        values = ['' for _ in keys]  #key리스트의 원소 수 만큼 ''문자를 원소를 갖는 value리스트를 만듬
        replaceDic = dict(zip(keys,values)) #key리스트 key로 value리스트를 value값으로 하는 사전 만듬

        return SearchEngine.replaceAll(text,replaceDic) #추출한 순수 텍스트에서 사전에 맵핑된 제거할 문자들을 모두 제거
    
    @staticmethod
    def freqOfWord(l): #단어의 빈도수를 사전으로 반환하는 함수
        word = []   
        for w in l: #리스트의 각 원소들이
            if (w in word) == False : word.append(w) #word리스트에 포함된 단어가 아니면 단어를 리스트에 추가 (단어중복제거)          
        freg = [l.count(w) for w in l] #리스트에서 단어의 출현 빈도수를 리스트로 만듬

        return dict(zip(word,freg)) #단어와 단어의 빈도수로 맵핑된 사전 반환
    
    @staticmethod
    def getKoreanAbb():  ##한국어 불용어 사전 불러오는 함수
        req = requests.get(' https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt')
        source = req.text
        return source.split()
    
    @staticmethod
    def getEnglishAbb(): ##영어 불용어 사전 불러오는 함수
        req = requests.get('http://www.lextek.com/manuals/onix/stopwords1.html')
        source = req.text

        ##정규표현식 활용 불용어 단어 파싱
        text = re.sub('\n',' ',source, 0, re.I|re.S)
        pattern = re.compile('<pre>(.+)</pre>')
        text = re.findall(pattern,text)[0]
        text = re.sub('#.*?#  #  #',' ',text, 0, re.I|re.S)
        return text.split()

    @staticmethod
    def filterAbbreviation(dic): ##사전에서 불용어를 제거하는 함수
        english_Abbreviation = SearchEngine.getEnglishAbb() ##불용어 사전 불러옴
        korean_Abbreviation = SearchEngine.getKoreanAbb()

        for key in list(dic.keys()):
            if key.lower() in english_Abbreviation : del dic[key]  ##영어 불용어 제거
            if key in korean_Abbreviation : del dic[key]        ##한국어 불용어 제거
        return dic
```

 ##### [8번 문제] 
 
 위 7번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 ***오버라이드***하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.
   - 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
   - getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.
     - reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
     - reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
     
   > \>\>\> w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
     
   > \>\>\> w4.getWordsFrequency()<br/>
   > [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]
     
   > \>\>\> w4.getWordsFrequency(reverse=true)<br/>
   > [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class SearchEngineWithOrderedWebWords(SearchEngine):
    def getWordsFrequency(self,reverse=False):   ## 기본인자는 False
        if self.urls == [] : return {}           ##등록된 URL이 없으면 빈 사전 반환
        if isinstance(reverse,bool) :            ##reverse에는 bool값인 True,False만 받을 수 있도록 한다.
            self.dicList = [] ##각 URL을 분석한 빈도수 사전을 저장할 리스트
            for url in self.urls:
                req = requests.get(url)
                text = SearchEngine.extractText(req.text)   #순수 텍스트 추출
                text = SearchEngine.filterSpecialChar(text) #특수문자 제거
                l = text.split()
                dic = SearchEngine.freqOfWord(l)           #출현한 단어의 빈도수 사전
                dic = SearchEngine.filterAbbreviation(dic) #사전 불용어 제거
                self.dicList.append(dic)                   #URL사전 저장

            self.totalDic = {}  ##종합 단어 출현빈도 사전
            for dic in self.dicList:            ##빈도사전을 하나씩 가져와서
                keys = list(dic.keys())         ##키를 리스트로 만들고
                for key in keys:                 
                    if key not in self.totalDic:     ##중복된 키가 없으면 사전에 추가
                        self.totalDic[key] = dic[key]

            self.items = list(self.totalDic.items()) ##단어 출현빈도 리스트
            if(reverse) : self.items.sort(key=lambda x: x[1], reverse=False) ##reverse가 true이면 오름차순 정렬 
            else : self.items.sort(key=lambda x: x[1], reverse=True) ##reverse가 false이면 내림차순 정렬
            return self.items
        
    def __iter__(self): 
        self.index = 0 ##단어 출현빈도 리스트를 반복할때 현재 index위치를 기억하고 있는 멤버
        return(self)
    
    def __next__(self):
        if self.index > (len(self.items)-1):  ##index 리스트의 범위를 초과하면
            raise StopIteration               ##예외발생
        item = self.items[self.index]  ##현재 index에 해당하는 리스트의 요소 --> 반환값
        self.index += 1                ##index를 리스트의 다음 요소 위치로 
        return item
```

```python
w1 = SearchEngineWithOrderedWebWords('http://www.catch.co.kr', 'http://www.naver.com', 'http://www.maxmovie.com')
print(w1.getWordsFrequency(reverse=True))
```
